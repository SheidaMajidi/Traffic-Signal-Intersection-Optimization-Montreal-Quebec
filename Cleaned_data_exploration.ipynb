## Finding intersections within 1 kilometer to each other

```python
# Define the Haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth's radius in kilometers
    lat1_rad, lon1_rad = np.radians(lat1), np.radians(lon1)
    lat2_rad, lon2_rad = np.radians(lat2), np.radians(lon2)
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance

# Function to find close intersections
def find_close_intersections(df, max_distance_km=1.0):
    # Extract unique intersections with their latitudes and longitudes
    unique_intersections = df[['Intersection_Name', 'Latitude', 'Longitude']].drop_duplicates()

    # List to store pairs of close intersections
    close_pairs = []

    # Iterate over each pair of intersections
    for i in range(len(unique_intersections)):
        for j in range(i + 1, len(unique_intersections)):
            # Get the latitudes and longitudes of the two intersections
            lat1, lon1 = unique_intersections.iloc[i]['Latitude'], unique_intersections.iloc[i]['Longitude']
            lat2, lon2 = unique_intersections.iloc[j]['Latitude'], unique_intersections.iloc[j]['Longitude']

            # Calculate the distance
            distance = haversine(lat1, lon1, lat2, lon2)

            # Check if the distance is within the threshold
            if distance <= max_distance_km:
                name1 = unique_intersections.iloc[i]['Intersection_Name']
                name2 = unique_intersections.iloc[j]['Intersection_Name']
                close_pairs.append((name1, name2, distance))

    return close_pairs

# Find intersections that are within 1 kilometer of each other
close_intersections = find_close_intersections(df, 1.0)

# Display the results
for pair in close_intersections:
    print(f"Intersections '{pair[0]}' and '{pair[1]}' are {pair[2]:.2f} km apart")
```

## Close intersections w data in the same day

```python
# Function to find close intersections with same-day data
def find_close_intersections_same_day(df, max_distance_km=1.0):
    # Extract unique intersections with their latitudes and longitudes
    unique_intersections = df[['Intersection_Name', 'Latitude', 'Longitude']].drop_duplicates()

    # List to store pairs of close intersections with same-day data
    close_pairs_same_day = []

    # Iterate over each pair of intersections
    for i in range(len(unique_intersections)):
        for j in range(i + 1, len(unique_intersections)):
            lat1, lon1 = unique_intersections.iloc[i]['Latitude'], unique_intersections.iloc[i]['Longitude']
            lat2, lon2 = unique_intersections.iloc[j]['Latitude'], unique_intersections.iloc[j]['Longitude']
            distance = haversine(lat1, lon1, lat2, lon2)

            if distance <= max_distance_km:
                name1 = unique_intersections.iloc[i]['Intersection_Name']
                name2 = unique_intersections.iloc[j]['Intersection_Name']

                # Check if there is same-day data for both intersections
                dates1 = set(df[df['Intersection_Name'] == name1]['Date'])
                dates2 = set(df[df['Intersection_Name'] == name2]['Date'])
                common_dates = dates1.intersection(dates2)

                if common_dates:
                    for date in common_dates:
                        close_pairs_same_day.append((name1, name2, date, distance))

    return close_pairs_same_day

# Find intersections that are within 1 kilometer of each other and have same-day data
close_intersections_same_day = find_close_intersections_same_day(df, 1.0)

# Display the results
for pair in close_intersections_same_day:
    print(f"Intersections '{pair[0]}' and '{pair[1]}' have data on {pair[2]} and are {pair[3]:.2f} km apart")
```

## Those close intersections that are in Sherbrooke

```python
def filter_intersections_by_name(intersections, name_keyword):
    # Filter for intersections with the specified keyword in their names
    filtered_intersections = [pair for pair in intersections if name_keyword in pair[0] or name_keyword in pair[1]]
    return filtered_intersections

# Filter the list to keep only intersections containing "Sherbrooke"
filtered_intersections = filter_intersections_by_name(close_intersections_same_day, "Sherbrooke")

# Display the results
for pair in filtered_intersections:
    print(f"Intersections '{pair[0]}' and '{pair[1]}' have data on {pair[2]} and are {pair[3]:.2f} km apart")
```

## Map these close intersections on Sherbrooke

```python
# Function to create a map with the filtered intersections
def create_map_with_intersections(df, filtered_intersections):
    # Initialize the map centered around the first intersection
    first_lat, first_long = df[df['Intersection_Name'] == filtered_intersections[0][0]][['Latitude', 'Longitude']].iloc[0]
    m = folium.Map(location=[first_lat, first_long], zoom_start=14)

    # Add markers for each intersection
    for pair in filtered_intersections:
        name1, name2 = pair[0], pair[1]

        lat1, long1 = df[df['Intersection_Name'] == name1][['Latitude', 'Longitude']].iloc[0]
        lat2, long2 = df[df['Intersection_Name'] == name2][['Latitude', 'Longitude']].iloc[0]

        folium.Marker([lat1, long1], popup=name1).add_to(m)
        folium.Marker([lat2, long2], popup=name2).add_to(m)

        # Draw lines between intersections
        folium.PolyLine([(lat1, long1), (lat2, long2)], color="blue").add_to(m)

    return m

map_with_intersections = create_map_with_intersections(df, filtered_intersections)

map_with_intersections
```

## Check and see if data is for weekday or weekend

```python
# Function to get the day type for the filtered intersections
def get_day_type_for_intersections(df, filtered_intersections):
    day_types = []

    # Iterate through the filtered intersections
    for pair in filtered_intersections:
        name1, name2, date, _ = pair

        # Get the Day_Type for the date of this pair
        day_type = df[(df['Intersection_Name'] == name1) & (df['Date'] == date)]['Day_Type'].iloc[0]
        day_types.append((name1, name2, date, day_type))

    return day_types

# Get the day types for the filtered intersections
day_types_for_intersections = get_day_type_for_intersections(df, filtered_intersections)

# Display the results
for info in day_types_for_intersections:
    print(f"On {info[2]}, intersections '{info[0]}' and '{info[1]}' were on a {info[3]}")
```

## Print in dataframe view

```python
def create_dataframe_for_intersections(df, filtered_intersections):
    # Create an empty list to store the data
    data = []

    # Iterate through the filtered intersections
    for pair in filtered_intersections:
        name1, name2, date, _ = pair

        # Get the Day_Type for the date of this pair
        day_type = df[(df['Intersection_Name'] == name1) & (df['Date'] == date)]['Day_Type'].iloc[0]
        data.append({'Intersection_1': name1, 'Intersection_2': name2, 'Date': date, 'Day_Type': day_type})

    # Create a DataFrame from the list
    df_result = pd.DataFrame(data)
    return df_result

# Create the DataFrame for the filtered intersections
df_intersections_day_type = create_dataframe_for_intersections(df, filtered_intersections)

# Display the DataFrame
df_intersections_day_type
```

```python
# Function to create a map with the intersections that have weekend data
def create_weekend_map(df, weekend_data):
    # Initialize the map centered around the first intersection in the weekend data
    first_lat, first_long = df[df['Intersection_Name'] == weekend_data.iloc[0]['Intersection_1']][['Latitude', 'Longitude']].iloc[0]
    m = folium.Map(location=[first_lat, first_long], zoom_start=14)

    # Add markers for each intersection in the weekend data
    for _, row in weekend_data.iterrows():
        name1, name2 = row['Intersection_1'], row['Intersection_2']

        lat1, long1 = df[df['Intersection_Name'] == name1][['Latitude', 'Longitude']].iloc[0]
        lat2, long2 = df[df['Intersection_Name'] == name2][['Latitude', 'Longitude']].iloc[0]

        folium.Marker([lat1, long1], popup=name1).add_to(m)
        folium.Marker([lat2, long2], popup=name2).add_to(m)

        # Optionally, draw lines between intersections
        folium.PolyLine([(lat1, long1), (lat2, long2)], color="blue").add_to(m)

    return m

weekend_map = create_weekend_map(df, weekend_data)

weekend_map
```
