## Finding intersections within 1 kilometer to each other

```python
# Define the Haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0  # Earth's radius in kilometers
    lat1_rad, lon1_rad = np.radians(lat1), np.radians(lon1)
    lat2_rad, lon2_rad = np.radians(lat2), np.radians(lon2)
    dlat = lat2_rad - lat1_rad
    dlon = lon2_rad - lon1_rad
    a = np.sin(dlat / 2)**2 + np.cos(lat1_rad) * np.cos(lat2_rad) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance

# Function to find close intersections
def find_close_intersections(df, max_distance_km=1.0):
    # Extract unique intersections with their latitudes and longitudes
    unique_intersections = df[['Intersection_Name', 'Latitude', 'Longitude']].drop_duplicates()

    # List to store pairs of close intersections
    close_pairs = []

    # Iterate over each pair of intersections
    for i in range(len(unique_intersections)):
        for j in range(i + 1, len(unique_intersections)):
            # Get the latitudes and longitudes of the two intersections
            lat1, lon1 = unique_intersections.iloc[i]['Latitude'], unique_intersections.iloc[i]['Longitude']
            lat2, lon2 = unique_intersections.iloc[j]['Latitude'], unique_intersections.iloc[j]['Longitude']

            # Calculate the distance
            distance = haversine(lat1, lon1, lat2, lon2)

            # Check if the distance is within the threshold
            if distance <= max_distance_km:
                name1 = unique_intersections.iloc[i]['Intersection_Name']
                name2 = unique_intersections.iloc[j]['Intersection_Name']
                close_pairs.append((name1, name2, distance))

    return close_pairs

# Find intersections that are within 1 kilometer of each other
close_intersections = find_close_intersections(df, 2.0)

# Display the results
for pair in close_intersections:
    print(f"Intersections '{pair[0]}' and '{pair[1]}' are {pair[2]:.2f} km apart")
```

## Close intersections w data in the same day

```python
# Function to find close intersections with same-day data
def find_close_intersections_same_day(df, max_distance_km=1.0):
    # Extract unique intersections with their latitudes and longitudes
    unique_intersections = df[['Intersection_Name', 'Latitude', 'Longitude']].drop_duplicates()

    # List to store pairs of close intersections with same-day data
    close_pairs_same_day = []

    # Iterate over each pair of intersections
    for i in range(len(unique_intersections)):
        for j in range(i + 1, len(unique_intersections)):
            lat1, lon1 = unique_intersections.iloc[i]['Latitude'], unique_intersections.iloc[i]['Longitude']
            lat2, lon2 = unique_intersections.iloc[j]['Latitude'], unique_intersections.iloc[j]['Longitude']
            distance = haversine(lat1, lon1, lat2, lon2)

            if distance <= max_distance_km:
                name1 = unique_intersections.iloc[i]['Intersection_Name']
                name2 = unique_intersections.iloc[j]['Intersection_Name']

                # Check if there is same-day data for both intersections
                dates1 = set(df[df['Intersection_Name'] == name1]['Date'])
                dates2 = set(df[df['Intersection_Name'] == name2]['Date'])
                common_dates = dates1.intersection(dates2)

                if common_dates:
                    for date in common_dates:
                        close_pairs_same_day.append((name1, name2, date, distance))

    return close_pairs_same_day

# Find intersections that are within 1 kilometer of each other and have same-day data
close_intersections_same_day = find_close_intersections_same_day(df, 1.0)

# Display the results
for pair in close_intersections_same_day:
    print(f"Intersections '{pair[0]}' and '{pair[1]}' have data on {pair[2]} and are {pair[3]:.2f} km apart")
```

# Minimizing average wait time at intersections

### Decision variables:

$$T_g $$  Duration of the green light (in seconds)

$$T_y $$  Duration of the yellow light (in seconds)

$$T_r $$  Duration of the red light (in seconds)

### Objective function:

Minimize: Average wait time for vehicles.

This can be represented as a function of the green light duration, the flow rate of vehicles, and the capacity of each lane. However, an exact formulation would require specific traffic flow models, which might make the problem nonlinear.

### Constraints:

#### 1. Traffic Light Cycle Constraint:

$$T_g + T_y + T_r = CycleLength $$

#### 2. inimum and Maximum Duration Constraints:

$$ T_{min-g} \leq T_g \leq T_{max-g} $$

$$ T_{min-y} \leq T_y \leq T_{max-y} $$

$$ T_{min-r} \leq T_r \leq T_{max-r} $$

#### 3. Vehicle Queue Constraint:

The number of vehicles that can queue at the intersection is limited by the length of the road leading up to it. This constraint can be represented by a function of the red light duration and the vehicle arrival rate.



```python
import gurobipy as gp
from gurobipy import GRB

try:
    # Create a new model
    m = gp.Model("traffic_light_optimization")

    # Define the parameters (adjust these based on your data and requirements)
    cycle_length = 120  # Total length of the traffic light cycle in seconds
    T_min_g, T_max_g = 10, 50  # Min and max duration for green light
    T_min_y, T_max_y = 3, 10   # Min and max duration for yellow light
    T_min_r, T_max_r = 10, 50  # Min and max duration for red light

    # Decision Variables
    T_g = m.addVar(lb=T_min_g, ub=T_max_g, vtype=GRB.CONTINUOUS, name="T_g")  # Green light duration
    T_y = m.addVar(lb=T_min_y, ub=T_max_y, vtype=GRB.CONTINUOUS, name="T_y")  # Yellow light duration
    T_r = m.addVar(lb=T_min_r, ub=T_max_r, vtype=GRB.CONTINUOUS, name="T_r")  # Red light duration

    # Objective Function: Minimize the average waiting time (simplified as maximizing green light time)
    # Note: For a more accurate model, you might need a more complex objective function
    m.setObjective(T_g, GRB.MAXIMIZE)

    # Constraints
    # 1. Traffic light cycle constraint
    m.addConstr(T_g + T_y + T_r == cycle_length, "cycle_constraint")

    # 2. Additional constraints can be added here based on your specific requirements
    # For example, vehicle queue constraints, safety constraints, etc.

    # Optimize the model
    m.optimize()

    # Print the optimal durations for each signal phase
    print(f"Optimal Green Light Duration: {T_g.X} seconds")
    print(f"Optimal Yellow Light Duration: {T_y.X} seconds")
    print(f"Optimal Red Light Duration: {T_r.X} seconds")

except gp.GurobiError as e:
    print('Error code ' + str(e.errno) + ": " + str(e))

except AttributeError:
    print('Encountered an attribute error')
```

```python
import gurobipy as gp
from gurobipy import GRB

# Define the optimization function
def optimize_traffic_lights(intersection_name, intersection_data):
    try:
        # Create a new model
        m = gp.Model("traffic_light_optimization_" + intersection_name)

        # Extract parameters from intersection_data
        arrival_rate = intersection_data['arrival_rate']
        current_timings = intersection_data['current_timings']
        
        # Define the parameters (you may need to adjust these based on your estimations)
        cycle_length = current_timings['Green'] + current_timings['Yellow'] + current_timings['Red']
        
        # Decision Variables
        T_g = m.addVar(lb=20, ub=60, vtype=GRB.CONTINUOUS, name="T_g")  # Green light duration
        T_y = m.addVar(lb=3, ub=10, vtype=GRB.CONTINUOUS, name="T_y")   # Yellow light duration

        # Red light duration is a dependent variable
        T_r = m.addVar(lb=0, ub=GRB.INFINITY, vtype=GRB.CONTINUOUS, name="T_r")  # Red light duration

        # Objective Function: Minimize the average waiting time
        # Here we use a simple proxy for the waiting time, which is the cycle length minus the green time
        m.setObjective(cycle_length - T_g, GRB.MINIMIZE)

        # Constraints
        # 1. Traffic light cycle constraint
        m.addConstr(T_g + T_y + T_r == cycle_length, "cycle_constraint")

        # 2. Additional constraints based on your requirements
        # Example: Minimum and maximum time for each signal phase

        # Optimize the model
        m.optimize()

        # Check if the model has been solved
        if m.status == GRB.OPTIMAL:
            return T_g.X, T_y.X, T_r.X
        else:
            print(f"Optimization was not successful for {intersection_name}. Status: {m.status}")
            return None

    except gp.GurobiError as e:
        print('Error code ' + str(e.errno) + ": " + str(e))
        return None
    except AttributeError as e:
        print('Encountered an attribute error:', e)
        return None

# Apply the optimization function to each of the specific intersections
for intersection in intersection_data:
    optimized_timings = optimize_traffic_lights(intersection, intersection_data[intersection])
    if optimized_timings is not None:
        print(f"Optimized timings for {intersection}: Green: {optimized_timings[0]}s, Yellow: {optimized_timings[1]}s, Red: {optimized_timings[2]}s")
    else:
        print(f"Optimization results not available for {intersection}.")
```
